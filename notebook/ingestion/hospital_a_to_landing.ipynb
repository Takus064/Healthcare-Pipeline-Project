```python
log_entries = []
# Function logging
def log_event(event_type, message, database = None, table = None):
    local_tz = timezone(timedelta(hours=7))
    log_entry = {
        'timestamp': datetime.now(timezone.utc).astimezone(local_tz).isoformat(),
        'event_type': event_type,
        'message': message,
        'database': database,
        'table': table
    }
    log_entries.append(log_entry)
    print(f"{log_entry['timestamp']} - {log_entry['event_type']}: {log_entry['message']} (Database: {log_entry['database']}, Table: {log_entry['table']})")
```

```python
# Hàm đọc file config từ Cloud Storage
def read_config_file():
    df = spark.read.format("csv").option("header", "true").load(config_file_path)
    log_event("CONFIG_READ", f"Config file read successfully")
    return df
```
